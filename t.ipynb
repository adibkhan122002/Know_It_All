```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('library.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Execute an SQL command to add a return_date column of type DATE to the borrows table
cursor.execute("ALTER TABLE members ADD COLUMN email TEXT;")

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('library.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Execute an SQL command to retrieve the schema of a table
cursor.execute("PRAGMA table_info(borrows);")

# Fetch the results and print them
table_schema = cursor.fetchall()
for column in table_schema:
    print(column)

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('library.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Execute the SQL query to retrieve table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")

# Fetch all the results as a list of tuples
table_names = cursor.fetchall()

# Print the table names
for name in table_names:
    print(name[0])

# Close the database connection
conn.close()
```
